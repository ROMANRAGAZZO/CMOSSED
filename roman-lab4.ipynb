## Questions:

1. We are taking more than a single flat-field exposure to increase the accuracy of the flat field correction with a median. Instead of a single exposure we are taking multiple and taking an average of them to reduce the random noise and uncertainties that come with taking a single image correction.

2. There are 2 reasons why the exposure times in the flats vary between filters. One reason is that different filters will allow different amounts of light through, so there will be differences in the amount of time needed to get the same number of photons on the detector with different filters. The second reason has to do with the detector's QE. The efficiency varies with wavelength, so the total time to get the same electron signal will also cause the exposure time to vary.

- Calculate a median-stacked bias exposure $\sf E_{bias, median}$ from your individual $\sf E_{bias}$ exposures.
    - Display the image $\sf E_{bias, median}$  with an informative color scale, reasonable limits, and labeling the units of your image.


```python
import numpy as np
from astropy.io import fits
from glob import glob
import matplotlib.pyplot as plt

# bias frames -> 103-107

bias_files = glob("/data/SBO-Data-Share/Classes/Fall-2024/ASTR3510/CMOSSED/2024-12-04/2024-12-04/Leto-Bias-*.fit")


E_bias = []
for i in range(len(bias_files)):
    E_bias.append(fits.open(bias_files[i])[0].data)
    
E_bias_median = np.median(E_bias, axis = 0)
```

```python
plt.figure(figsize = (10,8))
plt.imshow(E_bias_median,vmin = 58, vmax = 65)
plt.colorbar()
plt.title("Median Stacked Bias Image");
```


- Calculate a median-stacked dark exposure $\sf E_{dark, median}$ [ADU] from your individual $\sf E_{dark}$ exposures. Use it to calculate the rate at which dark current accumulates in each pixel $\sf \dot{N}_{dark}$.
    - Display this $\sf \dot{N}_{dark}$ image with an informative color scale, reasonable limits, and labeling the units of your image.

```python
dark_files = glob("/data/SBO-Data-Share/Classes/Fall-2024/ASTR3510/CMOSSED/2024-12-04/2024-12-04/Leto-Dark-10.000sec*.fit")

E_dark = []
for i in range(len(dark_files)):
    if (i == 0):
        Dark_0 = fits.open(dark_files[0])[0]
        darkheader = Dark_0.header
        E_dark.append(Dark_0.data)
        
    E_dark.append(fits.open(dark_files[i])[0].data)
    
E_dark_median = np.median(E_dark, axis = 0)

g = 4.01 # Pulled from slack and observing manual 
t_exp = 10

N_dark = g*(E_dark_median - E_bias_median)/t_exp
```

```python
plt.figure(figsize = (10,8))
plt.imshow(N_dark, vmin = -0.50, vmax = 0.5)
plt.colorbar()
plt.title("$N_{dark}$ Calibrated Image (e-)");
```

- Calculate a median-stacked flat-field exposure $\sf E_{flat, median}$ [ADU] from your individual $\sf E_{flat}$ exposures. Use it to calculate a normalized flat-field image $\sf I_{flat-field}$. Create a flat-field image for *each* filter.
    - Display your $\sf I_{flat-field}$ image(s) with an informative color scale and reasonable limits.
    - Qualitatively describe the flat-field calibration images. What features do you see? By approximately what percent does the sensitivity of your telescope + detector + filter setup vary across the field of view? How are they different across the three filters?

```python
flat_files_i = glob("/data/SBO-Data-Share/Classes/Fall-2024/ASTR3510/CMOSSED/2024-12-04/2024-12-04/Leto-FlatField-35.000sec*.fit")
flat_files_g = glob("/data/SBO-Data-Share/Classes/Fall-2024/ASTR3510/CMOSSED/2024-12-04/2024-12-04/Leto-FlatField-2.000sec*.fit")
flat_files_r = glob("/data/SBO-Data-Share/Classes/Fall-2024/ASTR3510/CMOSSED/2024-12-04/2024-12-04/Leto-FlatField-1.500sec*.fit")

E_flat_i = []
E_flat_g = []
E_flat_r = []
for i in range(len(flat_files_i)):
    E_flat_i.append(fits.open(flat_files_i[i])[0].data)
    E_flat_g.append(fits.open(flat_files_g[i])[0].data)
    E_flat_r.append(fits.open(flat_files_r[i])[0].data)
    
E_flat_median_i = np.median(E_flat_i, axis = 0)
t_flat_i = 35
I_flat_i = (E_flat_median_i - E_bias_median - (N_dark*t_flat_i)/g)/np.median(E_flat_median_i - E_bias_median - (N_dark*t_flat_i)/g)

E_flat_median_g = np.median(E_flat_g, axis = 0)
t_flat_g = 2.0
I_flat_g = (E_flat_median_g - E_bias_median - (N_dark*t_flat_g)/g)/np.median(E_flat_median_g - E_bias_median - (N_dark*t_flat_g)/g)

E_flat_median_r = np.median(E_flat_r, axis = 0)
t_flat_r = 1.5
I_flat_r = (E_flat_median_r - E_bias_median - (N_dark*t_flat_r)/g)/np.median(E_flat_median_r - E_bias_median - (N_dark*t_flat_r)/g)
```

```python
fig, ax = plt.subplots(3,1,figsize = (10,8))
pcm1 = ax[0].imshow(I_flat_g)
pcm2 = ax[1].imshow(I_flat_r)
pcm3 = ax[2].imshow(I_flat_i, vmin = 0.95)
fig.colorbar(pcm1, shrink = 0.8);
fig.colorbar(pcm2, shrink = 0.8)
fig.colorbar(pcm3, shrink = 0.8)
ax[0].set_title("Calibrated Flat Field Images (g',r',i' respectively)");
```

If the sky is clear, the effects of Earth’s atmospheric transmission for a particular filter can be modeled by the linear function $\sf m_{telescope} = m_{space} + k_λX$, where $m_{telescope}$ is the magnitude that reaches the telescope, $\sf m_{space}$ is the magnitude of the light source above Earth’s atmosphere, $\sf k_λ$ is the first-order extinction coefficient, and $\sf X$ is the airmass. Extinction coefficients for SBO are approximately:


| Filter                 | ${k}_{\lambda}$ |
|:------------------:|:-------------:|
| g'              | $\sf 0.21 ~^{mag}/_{airmass}$|
| r'              | $\sf 0.13 ~^{mag}/_{airmass}$|
| i'              | $\sf 0.15 ~^{mag}/_{airmass}$|
