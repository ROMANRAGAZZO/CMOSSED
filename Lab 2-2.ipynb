Our detector has pixel lengths to 1600 and 2400. If we divide that by plate scale it will give us angular size, or FOV

```python
yFOV = 1600/plate_scale_est
```

```python
xFOV = 2400/plate_scale_est
```

```python
print(f"This is how big the vertical side is: {yFOV/60} arcminutes")
print(f"This is how big the horizontal side is: {xFOV/60} arcminutes")
```
